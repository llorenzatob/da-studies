## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

````
# (A)
qtd_total_vendas = 3
tkt_medio = 320.52
A = qtd_total_vendas + tkt_medio
print(A)
````

````
# (B)
valor_total_vendas = 834.47
tkt_medio = 119.21
B = valor_total_vendas // tkt_medio
print(B)
````

````
# (C)
valor_total_vendas = 15378.12
qt_total_vendas = 5
C = valor_total_vendas / qt_total_vendas
print(C)
````

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

````
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

contagem = cancao.count("roda")

print(contagem)

capslock = cancao.upper()

print(capslock)

swap = cancao.title()

print(swap)


````

## 4\. Booleanos

Utilize a tabela da verdade para responder: qual o valor da variável x?

````
a = False
b = True

x = not a & b

print(False & True)
print(x)
````

Sem o ```not ``` a variável ```x``` seria ```False```, mas como tem essa operação lógica correlacionada, o valor é ```True```.
